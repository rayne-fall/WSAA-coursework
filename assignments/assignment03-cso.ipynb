This notebook contains a Python program designed to retrieve the historical data related to the exchequer account from the Central Statistics Office and save it to a local file.

## Imports

```python
import requests
import json
```

## Retrieving the Data

```python
# identify url
url="https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/2.0/en"
# get the dataset
exchequer_data=requests.get(url)
```

## Saving the Data

```python
# create cso.json file and save data to it [1]
with open("cso.json","w") as outfile:
    json.dump(exchequer_data.text, outfile)
```
